```python
cursor.execute("SELECT Valor FROM Mediciones WHERE Tipo like 'Altura'")
altura=cursor.fetchall()
```

```python
sns.distplot(altura,color='y')
```

La altura parece seguir una distribución normal en los individuos muestreados, con una media entre 1.6 y 1.7 m.

```python
cursor.execute("SELECT Valor FROM Mediciones WHERE Tipo like 'Frente'")
frente=cursor.fetchall()
sns.distplot(frente,color='b')
```

A diferencia de la altura, podemos observar que la varianza es mayor en las medidas de pigmentación en la frente (un área del cuerpo que, en la mayoría de las personas, está expuesta a la radiación solar con frecuencia), quizás está diferencia se debe a que las personas muestreadas tienen diferente duración en tiempos de exposición al sol, donde las personas con un nivel de pigmentación mayor serían quienes se exponen a la radiación solar más frecuentemente o porque tienen tiempos de exposiciń más largos; sin embargo, esta medida no refleja totalmente la influencia de la exposición solar en la pigmentación de la piel, ya que esta variable no considera otros factores como el uso de maquillaje en la medición, grasa en la cara, etc.

```python
cursor.execute("SELECT Valor FROM Mediciones WHERE Tipo like 'Brazo'")
brazo=cursor.fetchall()
sns.distplot(brazo,color='g')
```

En cuanto a pigmentación en la parte interior del brazo, la cual se supone que es un área del cuerpo con poca exposición a la luz solar, observamos un comportamiento menos drástico que con la pigmentación en la frente, una posible explicación sería que la pigmentación en el brazo refleja una medida mayoritariamente afectada por factores genéticos (bajo el supuesto de que esta área del cuerpo no está expuesta al sol tan frecuentemente) y que, entonces, este comportamiento mas "normal" es explicado porque la población muestrada pertenece a una misma población mexicana.
